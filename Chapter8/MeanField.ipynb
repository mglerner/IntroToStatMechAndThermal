```python
import numpy as np, scipy as sp
from matplotlib import pyplot as plt
import seaborn as sns
from ipywidgets import interact,fixed
```

```python
def plot_s(ben=0):
    fig,ax = plt.subplots()
    s = np.linspace(-2,2,100)
    ax.plot(s,s,label=r'$\bar{s}$')
    ax.plot(s,np.tanh(ben*s),label=r'$\tanh(\beta \epsilon n \bar{s})$')
    ax.grid('on')
    ax.set_xlabel(r'$\bar{s}$')
    ax.set_title(rf'$\beta \epsilon n = {ben:.3f}$')
    ax.legend()
interact(plot_s,ben=(0,4,0.1))
```
