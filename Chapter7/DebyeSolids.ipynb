```python
import numpy as np, scipy as sp
from matplotlib import pyplot as plt
import seaborn as sns
from ipywidgets import interact, fixed
```

```python
eokt = np.linspace(0.0001,4,1000)
def cv_einstein(eokt):
    top = eokt**2 * np.exp(eokt)
    bottom = (np.exp(eokt) - 1)**2
    return top/bottom
plt.plot(eokt,cv_einstein(eokt))
```

```python
eokt = np.linspace(0.0001,1,1000)
plt.plot(eokt,eokt**2,'k--',label='1')
plt.plot(eokt,np.exp(eokt),label='2')
plt.plot(eokt,(np.exp(eokt) - 1)**2,label='3')
plt.legend()
```
