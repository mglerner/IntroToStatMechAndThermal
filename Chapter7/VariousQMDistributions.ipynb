We know the fermi dirac distribution looks like

$$ \bar{n}_s = \frac{1}{e^{\beta(\epsilon_s - \mu)}+1}$$


We can rewrite that to make it a bit more generic to plot. We'll want $\bar{n}$ as a function of $\epsilon_s/\mu$ to scale the energy. Then we want a range of temperatures that show the full behavior. That means we want a range of $\beta$. It turns out that, in terms of $\mu$, a good range to look it as letting $k_BT$ go from $\mu/1000$ all the way up to $\mu$. Or, we could say that $k_BT/\mu$ goes from 1/1000 to 1. So, if we rewrite things a bit, we get

$$n_s = \frac{1}{e^{\left(\frac{1}{k_BT/\mu}\right)(\epsilon_s/\mu - 1)}}$$

```python
from numpy import exp, linspace, logspace
from matplotlib import pyplot as plt
from ipywidgets import interact
```

```python
kB = 1.380649e-23 # m2 kg s-2 K-1
def _ns(T,epsilons,mu):
    beta = 1/(kB*T)
    return 1/(exp(beta*epsilons-mu)+1)

def ns(kbTom,esom):
    #kbTom means kbT over mu; esom means epsilon_s / mu
    exponent = (1/kbTom)*(esom - 1)
    return 1/(exp(exponent) + 1)
```

```python
esom = linspace(0,5,1000)
plt.plot(esom,ns(1/1000,esom))
```

```python
def plot_ns(kbTom=0.001):
    esom = linspace(0,5,1000)
    plt.plot(esom,ns(kbTom,esom))
    plt.title(f'$k_B T/\mu$ = {kbTom}')
    plt.xlabel('$\epsilon_s/\mu$')
    plt.ylabel('$n_s$')
interact(plot_ns,kbTom=(0.001,1,0.001));
```
